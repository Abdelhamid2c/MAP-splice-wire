```python
import win32com.client
import os

def dupliquer_feuille_template(chemin_fichier, nom_template, nombre_copies, poste="poste_"):

    
    chemin_absolu = os.path.abspath(chemin_fichier)
    print(f"Ouverture du fichier {chemin_absolu}...")
    
    # Créer une instance d'Excel
    excel = win32com.client.Dispatch("Excel.Application")
    excel.Visible = False  # Ne pas afficher Excel pendant le processus
    excel.DisplayAlerts = False  # Désactiver les alertes
    
    try:
        # Ouvrir le classeur
        classeur = excel.Workbooks.Open(chemin_absolu)
        
        # Vérifier si la feuille template existe
        feuille_existe = False
        for feuille in classeur.Sheets:
            if feuille.Name == nom_template:
                feuille_existe = True
                feuille_template = feuille
                break
        
        if not feuille_existe:
            print(f"Erreur: La feuille '{nom_template}' n'existe pas dans ce classeur.")
            feuilles_dispo = [feuille.Name for feuille in classeur.Sheets]
            print(f"Feuilles disponibles: {', '.join(feuilles_dispo)}")
            classeur.Close(SaveChanges=False)
            excel.Quit()
            return
        
        print(f"Feuille template '{nom_template}' trouvée.")
        
        # Créer les copies
        for i in range(1, nombre_copies + 1):
            nouveau_nom = f"Diversité_{i}"
            
            # Vérifier si une feuille avec ce nom existe déjà
            feuille_existe = False
            for feuille in classeur.Sheets:
                if feuille.Name == nouveau_nom:
                    feuille_existe = True
                    break
            
            if feuille_existe:
                print(f"Attention: Une feuille nommée '{nouveau_nom}' existe déjà et sera ignorée.")
                continue
            
            print(f"Création de la feuille '{nouveau_nom}'...")
            # Copier la feuille (After=-1 signifie à la fin)
            feuille_template.Copy(After=classeur.Sheets(classeur.Sheets.Count))
            # La nouvelle feuille est maintenant la dernière feuille
            nouvelle_feuille = classeur.Sheets(classeur.Sheets.Count)
            nouvelle_feuille.Name = nouveau_nom
        
        feuille_template.Delete()
        # Enregistrer sous un nouveau nom
        dossier = os.path.dirname(chemin_absolu)
        # nom_base = os.path.splitext(os.path.basename(chemin_absolu))[0]
        extension = os.path.splitext(chemin_absolu)[1]
        nouveau_chemin = os.path.join(dossier, f"{poste}_MAP{extension}")
        
        print(f"Enregistrement des modifications dans {nouveau_chemin}...")
        classeur.SaveAs(nouveau_chemin)
        classeur.Close(SaveChanges=True)
        
        print(f"Terminé! {nombre_copies} copies ont été créées avec succès.")
        print(f"Nouveau fichier enregistré: {nouveau_chemin}")
        
    except Exception as e:
        print(f"Erreur: {e}")
    finally:
        # Fermer Excel
        excel.Quit()
        return nouveau_chemin
              
    
# dupliquer_feuille_template(chemin_fichier, nom_template, nombre_copies, prefixe_nom)
```

```python
color_dict = {
    'Y': 'FFFF00',  # Yellow
    'G': '00FF00',  # Green
    'L': '0000FF',  # Blue
    'R': 'FF0000',  # Red
    'W': 'FFFFFF',  # White
    'LG': '66FF00', # Gray (Light Gray)
    'O': 'FFA500',  # Orange
    'BR': '80471c',#'A52A2A', # Brown
    'V': '800080',#'8A2BE2',   # Violet
    'GY': '707070',#'C0C0C0',  # Gray
    'B': '000000' ,  # Black (Noir)
    'P' : 'FFC0CB', # Pink
    'C' : '00FFFF', # Cyan
    'D' : 'FFFFF0', # Ivory
    'SI' : '505050', # Silver
    'DL' : '0000FF'
}

def apply_color(symbol, cell):
    if '/' in str(symbol):
        primary, secondary = symbol.split('/')
        fill = PatternFill(start_color=color_dict[primary], end_color=color_dict[primary], fill_type="solid")
        cell.fill = fill
        
        print(color_dict[secondary])
        
        side = Side(border_style="thick", color=color_dict[secondary])
        border = Border(diagonal=side, diagonalUp=True)
        cell.border = border
    elif symbol in color_dict:
        fill = PatternFill(start_color=color_dict[symbol], end_color=color_dict[symbol], fill_type="solid")
        cell.fill = fill
    else :
        cell.value = symbol
    
```
