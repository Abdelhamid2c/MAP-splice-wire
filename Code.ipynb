```python
import pandas as pd
pd.set_option('display.max_columns', None)

df = pd.read_excel(path_data, sheet_name="Principal", )
```

```python
PN_columns = df.columns.tolist()[18:]
```

```python
df_sm4 = df[df['POSTE']== 'P2']
```

```python
import pandas as pd
import numpy as np
from collections import defaultdict

# Créer un dictionnaire pour stocker les colonnes identiques pour chaque colonne
identical_columns = defaultdict(list)

# Pour chaque colonne dans PN_columns
for col1 in PN_columns:
    identical_columns[col1] = []
    
    # Comparer avec toutes les autres colonnes
    for col2 in PN_columns:
        if col1 != col2:
            # Vérifier si les colonnes sont identiques (en ignorant les NaN)
            is_identical = df_sm4[col1].equals(df_sm4[col2]) or df_sm4[col1].fillna('NaN').equals(df_sm4[col2].fillna('NaN'))
            if is_identical:
                identical_columns[col1].append(col2)

# Identifier les groupes de colonnes identiques
column_groups = []
processed_columns = set()

for col in PN_columns:
    if col not in processed_columns:
        # Créer un nouveau groupe avec cette colonne
        group = [col] + identical_columns[col]
        
        # Marquer toutes les colonnes de ce groupe comme traitées
        processed_columns.update(group)
        
        # Ajouter le groupe à la liste des groupes
        column_groups.append(group)

# Afficher les résultats
print(f"Nombre total de groupes: {len(column_groups)}")
for i, group in enumerate(column_groups, 1):
    print(f"\nGroupe {i} ({len(group)} colonnes):")
    print(", ".join(group))
    
    if len(group) > 1:
        sample_values = df_sm4[group[0]].dropna().unique()[:5]  # Afficher jusqu'à 5 valeurs uniques
        print(f"Exemples de valeurs: {sample_values}")
```

```python
# Choisir le groupe pour lequel vous voulez extraire les valeurs uniques de Splice
# Par exemple, pour le premier groupe:
group_index = 0  # Changez cet index pour sélectionner un autre groupe

if group_index < len(column_groups) and len(column_groups) > 0:
    selected_group = column_groups[group_index]
    print(f"Groupe sélectionné: {', '.join(selected_group)}")
    
    # Prendre la première colonne du groupe comme référence
    reference_column = selected_group[0]
    
    # Trouver les lignes où Splice est présent dans cette colonne
    mask = df_sm4[reference_column].notna()
    
    # Obtenir les valeurs uniques de Splice pour ce groupe
    unique_splices = df_sm4.loc[mask, 'Splice'].unique()
    
    print(f"\nValeurs uniques de SPLICE pour ce groupe ({len(unique_splices)} valeurs):")
    for i, splice in enumerate(unique_splices, 1):
        print(f"{i}. {splice}")
    
```

```python
# Utiliser le groupe 1 (indice 0) 
group_index = 0  

if group_index < len(column_groups) and len(column_groups) > 0:
    selected_group = column_groups[group_index]
    print(selected_group)
#     # Prendre la première colonne du groupe comme référence
#     reference_column = selected_group[0]
    
#     # Filtrer le DataFrame où la colonne de référence est égale à 1.0
    df_group1_eq1 = df_sm4[df_sm4[reference_column] == 1.0]
    
```

```python
for splice in df_group1_eq1['Splice'].unique():
        print(df_group1_eq1[df_group1_eq1['Splice'] == splice])
```
